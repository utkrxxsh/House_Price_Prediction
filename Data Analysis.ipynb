```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
print(dataset.shape)
```

```python
dataset.head()
```

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
```

```python
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing values')
```

```python
for feature in features_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numerical_variables=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
print('Number if Numerical Variables is {}'.format(len(numerical_variables)))
dataset[numerical_variables].head()
```

```python
year_feature=[feature for feature in numerical_variables if 'Yr' in feature or 'Year' in feature]
```

```python
print(len(year_feature))
year_feature
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('House Price')
plt.title('House Price vs Year Sold')
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('House Price')
        plt.show()
```

```python
discrete_features=[feature for feature in numerical_variables if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
```

```python
print(len(discrete_features))
```

```python
discrete_features
```

```python
dataset[discrete_features].head()
```

```python
for feature in discrete_features:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('House Price')
    plt.title(feature)
    plt.show()
```

```python
continous_features=[feature for feature in numerical_variables if feature not in discrete_features+year_feature+['Id']]
print(len(continous_features))
continous_features
```

```python
for feature in continous_features:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
for feature in continous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('House Price')
        plt.title(feature)
        plt.show()
```

```python
for feature in continous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
print(len(categorical_features))
```

```python
dataset[categorical_features].head()
```

```python
for feature in categorical_features:
    print('Number of Unique values in {} is {}'.format(feature,len(dataset[feature].unique())))
```

```python
for feature in categorical_features:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('House Price')
    plt.title(feature)
    plt.show()
```
